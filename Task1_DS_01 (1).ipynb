## Data Visualization + EDA

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
```

```python
sns.histplot(df['2022 Population'],bins=10,kde=True)
plt.title('Distribution of 2022 Population')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```

```python
df
```

```python
ddf=df.groupby('Continent')['2022 Population'].mean().reset_index()
```

```python
ddf
```

```python
top10_countries_by_population=df.sort_values(by='2022 Population',ascending=False).head(10)
```

```python
sns.barplot(x='2022 Population',y='Country/Territory',data=top10_countries_by_population)
plt.xlabel='2022 Population'
plt.ylabel='Country'
plt.title('Top 10 Countries by Population in 2022')
plt.show()
```

```python
sns.barplot(x='Continent',y='Density (per km²)',data=df)
plt.title('Population density of different Continents')
plt.show()
```

```python
sns.boxplot(x='Continent',y='Density (per km²)',data=df,showfliers=False)
plt.title('Population density distribution')
```
